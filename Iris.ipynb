```python
#import libraries
import numpy as np
import pandas as pd 
from math import sqrt
```

```python
Iris = pd.read_csv('Iris.csv') # load the csv file
Iris.head()
```

```python
Iris=Iris.sample(frac=1).reset_index(drop=True) # Shuffle the dataset because our dataset was groupped according to flower name
```

```python
Iris.head()
```

```python
training = int(Iris.shape[0]*0.7) # Set the 70% of dataset as train and 30% of dataset as test
X_Train = Iris[:training] #70% is train
X_Test = Iris[-(Iris.shape[0]-training):] #30% is test
```

```python
my_train = X_Train.values # set my train and test variables 
my_test = X_Test.values
Y_Train = my_test[:,-1]
```

```python
print('Shape of Train: ',X_Train.shape)
print('Shape of Test : ',X_Test.shape)
```

```python
def euclidean_distance(x_train,x_test): # calculate the distance between two points using euclidean method
    dist = 0 # in the beginnig distance is 0
    for i in range(len(x_test)-1): # travell all the test elements
        dist += (x_test[i]-x_train[i])**2 # use euclidean  distance method
    return sqrt(dist) # return the result
```

```python
def List_neighbors(x_train,x_test,k): # get the nearest k neighbors 
    model = [] #data list
    distances = [] # distances list
    for i in x_train: # travell all x_train elements
        distances.append(euclidean_distance(i,x_test)) # get the distance between x_test and i items
        model.append(i) # append the i  in the data array
    distances = np.array(distances) # set the distances array as numpy array
    model = np.array(model) # set the data array as numpy array
    sort_indexes = distances.argsort() #sort the distances that in the distances array ascendant
    model = model[sort_indexes]
    return model[:k]  # return first k nearest neighbors
```

```python
def Predict(x_test, x_train, k): # Predict function according to k values
    data = [] # my data
    neighbors = List_neighbors(x_train,x_test,k) # get neighbors
    for i in neighbors: #travell neighbors
        data.append(i[-1])
    result=pred_percentage(data,k)# get the prediction and prediction percentage
    return result
```

```python
def accuracy(species, y_test): # calculate the accuracy
    num_of_correct = 0 # number of correct prediction
    for i in range(len(species)): # start travelling
        if species[i]==y_test[i][0]: # if prediction is true
            num_of_correct+=1 # count pred
    accuracy = num_of_correct/len(species) # calculate the rate of true predictions
    return accuracy # return the result
```

```python
def pred_percentage(classes,k):
    seto=0 # set the numbers of species as 0 
    vers=0
    virg=0
    t=0
    for i in classes: # start travelling 
        if i=='Iris-setosa':
            seto+=1 
        elif i=="Iris-versicolor":
            vers+=1
        elif i=="Iris-virginica":
            virg+=1
        t+=1
        if t==k:
            summary=seto+vers+virg
            seto=round(seto/summary,3) #calculate the ratio of each
            vers=round(vers/summary,3)
            virg=round(virg/summary,3)
            var = {seto:"Iris-setosa",vers:"Iris-versicolor",virg:"Iris-virginica"}
            pred=var.get(max(var)) #get the majority
            pred_per=max(var)*100  # calculate percentage
            print("Setosa: ",seto," Versicolor: ",vers," Virginica: ", virg,"****  Prediction: ",pred," %",pred_per," For k=",k)
            result=[pred,pred_per] # set result list
            return result # return calculated result
```

```python
pred1 = [] # prediction for k=1
pred3 = [] # prediction for k=3
pred5 = [] # prediction for k=5
for i in my_test:
    pred1.append(Predict(i,my_train, 1))
    pred3.append(Predict(i,my_train, 3))
    pred5.append(Predict(i,my_train, 5))
    print("***")
```

```python
accuracy1 = accuracy(Y_Train, pred1) # accuracy for k=1
accuracy3 = accuracy(Y_Train, pred3) # accuracy for k=3
accuracy5 = accuracy(Y_Train, pred5) # accuracy for k=5
print("Accuracy for k=1: ",accuracy1,"\nAccuracy for k=3: ",accuracy3,"\nAccuracy for k=5: ",accuracy5) 
```
